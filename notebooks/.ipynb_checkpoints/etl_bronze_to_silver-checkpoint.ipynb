```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, trim

# Khởi tạo SparkSession
spark = SparkSession.builder \
    .appName("Transfermarkt ETL") \
    .getOrCreate()

print("Spark session started")

# Đọc dữ liệu từ lớp Bronze (data_raw)
df_players = spark.read.csv(
    "/home/jovyan/data_raw/players.csv",
    header=True,
    inferSchema=True
)

print("Số dòng ban đầu:", df_players.count())
df_players.printSchema()

# Làm sạch dữ liệu
df_players_clean = (
    df_players
    .dropna(subset=["player_name"])                      # Xóa dòng thiếu tên
    .withColumnRenamed("player_name", "name")            # Đổi tên cột
    .withColumn("age", col("age").cast("int"))           # Chuyển kiểu dữ liệu
    .withColumn("name", trim(col("name")))               # Xóa khoảng trắng
)

print("Số dòng sau khi làm sạch:", df_players_clean.count())

# Ghi dữ liệu sang Silver Layer
output_path = "/home/jovyan/work/silver/players"

df_players_clean.write.mode("overwrite").parquet(output_path)
print(f"Dữ liệu Silver đã được ghi thành công tại: {output_path}")

# Kiểm tra lại dữ liệu Silver
df_check = spark.read.parquet(output_path)
df_check.show(5)
```
